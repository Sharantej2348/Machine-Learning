# Merge

```python
var1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})

var2=pd.DataFrame({"A":[1,2,3,4],"C":[21,22,23,24]})

pd.merge(var1,var2,on="A")
```

```python
var1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})

var2=pd.DataFrame({"A":[1,2,3,5],"C":[21,22,23,24]})

pd.merge(var1,var2,how="left",indicator=True)

# how can be left,right,outer,inner
```

```python
var1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})

var2=pd.DataFrame({"A":[1,2,3,5],"B":[21,22,23,24]})

pd.merge(var1,var2,left_index=True,right_index=True,suffixes=("name","id"))
```

# Concat

```python
sr1 = pd.Series([1,2,3,4])

sr2 = pd.Series([11,12,13,14])

pd.concat([sr1,sr2])
```

```python
d1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})

d2=pd.DataFrame({"A":[1,2,3,5],"C":[21,22,23,24]})

pd.concat([d1,d2],axis=1)
```

```python
d1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})

d2=pd.DataFrame({"A":[1,2],"C":[21,22]})

pd.concat([d1,d2],axis=1,join="inner")
```

```python
d1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})

d2=pd.DataFrame({"A":[1,2],"C":[21,22]})

pd.concat([d1,d2],axis=1,keys=["d1","d2"])
```

# Join

```python
import pandas as pd
```

```python
var1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]},index=["a","b","c","d"])

var2=pd.DataFrame({"C":[10,20],"D":[11,22]},index=["a","b"])

var1.join(var2)
```

```python
var1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]},index=["a","b","c","d"])

var2=pd.DataFrame({"C":[10,20],"D":[11,22]},index=["a","b"])

var2.join(var1,how = "right")
```

```python
var1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]},index=["a","b","c","d"])

var2=pd.DataFrame({"C":[10,20],"B":[11,22]},index=["a","b"])

var2.join(var1,how = "right",lsuffix="_12",rsuffix="_13")
```

# pivot()

```python
var = pd.DataFrame({"days" : [1,2,3,4,5,6],"std_name":["a","b","a","a","b","b"],"eng" : [10,12,14,15,16,12],"maths" : [17,18,19,13,14,16]})

var
```

```python
var.pivot(index="days",columns="std_name",values="eng")
```

```python
var.pivot(index="days",columns="std_name")
```

```python
piv = pd.DataFrame({"days" : [1,1,1,1,2,2],"std_name":["a","b","a","a","b","b"],"eng" : [10,12,14,15,16,12],"maths" : [17,18,19,13,14,16]})

piv
```

```python
piv.pivot_table(index="std_name",columns="days",
                aggfunc="mean")
```

```python
piv.pivot_table(index="std_name",columns="days",
                aggfunc="sum",margins="True")
```
