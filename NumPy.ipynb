# Join Array

```python
var1 = np.array([1,2,3,4])

var2 = np.array([9,8,7,6])

ar = np.concatenate((var1,var2))

print(ar)
```

```python
var1 = np.array([[1,2],[3,4]])

var2 = np.array([[9,8],[7,6]])

ar = np.concatenate((var1,var2),axis=1)


print(ar)
```

```python
import numpy as np

var1 = np.array([1,2,3,4])

var2 = np.array([9,8,7,6])

ar = np.stack((var1,var2),axis=1)

ar_1 = np.hstack((var1,var2))      #row

ar_2 = np.vstack((var1,var2))      #column

ar_3 = np.dstack((var1,var2))      #height

print(ar)
print()
print(ar_1)
print()
print(ar_2)
print()
print(ar_3)
```
